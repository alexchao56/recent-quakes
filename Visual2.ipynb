```python
import urllib
from pandas import read_csv

url= 'http://earthquake.usgs.gov/earthquakes/catalogs/eqs7day-M1.txt'
data= read_csv(urllib.urlopen(url))
#dont need urllib. if said from urllib import urlopen

data[0:5]

clean_data = data.dropna(axis=0,how='any')
#pandas.Data Frame.dropna; axis:0-row, 1-col; how:want to drop the row that has any NA's;
#naming differently, better refering to specific dataframe, python allows running codes in any order any numbers of time;
#this 'clean_data' is only changing 'data';
#so no matter how many times you run this code, as long as 'data' is unchanged, 'clean_data' will always be the same.
clean_data[0:3]
alaska = clean_data[clean_data.Src == 'ci']
#clean_data.Src=='ak' tells which rows in the Src column have value =='ak' 
#clean_data[clean_data.Src=='ak'] :subsetting those rows
alaska[0:10]
```

```python
from mpl_toolkits.basemap import Basemap

def G1_quakes(abbr):      #abr=abbreviation of the state
    state = clean_data[clean_data.Src == abbr]     #subseting the earthquakes in a specific state
    
    #setting coordinates for displaying the state:
    lllon=-(min(max(abs(state.Lon))+5,180))          
        #there has been a problem of simply using the minimum of the logitudes for Alaska because the eastmost location has 
        #logitude on the east atmosphere (E 173), which is positive whereas all locations of other states are negative.
        #to solve this, take the max of the absolute values of Lon. the times '-1', also dont want it to exceed -180.
    lllat=min(state.Lat)-5
    urlon=-(min(abs(state.Lon))-5)
    urlat=max(state.Lat)+5
    
    #Basemap:
    fig = plt.figure(figsize=(8,8))
    m = Basemap(llcrnrlon=lllon, llcrnrlat=lllat,
                urcrnrlon=urlon,urcrnrlat=urlat,
                resolution='l', area_thresh=1000.,projection='merc')
    m.drawcoastlines()
    m.drawcountries()
    m.drawstates()
    m.fillcontinents(color='coral', lake_color='aqua')
    m.drawmapboundary(fill_color='aqua')
    
    #plotting the earthquakes as dots with different sizes proportional to magnitudes:
    lon = state.Lon  
    lat = state.Lat
    xpt,ypt = m(lon,lat)
    lonpt, latpt = m(xpt,ypt,inverse=True)
    sizes=state.Magnitude
    for i in range(1,len(lon)):
        m.plot(xpt[i-1:i],ypt[i-1:i],'ro',markersize=sizes[i-1:i]*5,alpha=0.7)
    plt.show()
```
